```python
#%%

import pandas as pd
import os

data_dir = '/Users/kshitijmac/Documents/Data_Files/retail_db'

# Dictionary to hold DataFrames
dataframes = {}

for table in os.listdir(data_dir):
    table_folder = os.path.join(data_dir, table)

    # Skip non-directory entries
    if not os.path.isdir(table_folder):
        continue

    file_path = os.path.join(table_folder, 'part-00000')
    
    if os.path.exists(file_path):
        print(f"Reading {table} from {file_path}")
        
        # Try to read assuming tab-separated (common in part-00000 files)
        df = pd.read_csv(file_path, header=None, delimiter='\t')
        dataframes[table] = df

# Example usage:
print(dataframes['categories'].head())
```
